### importing covid19 dataset
importing "Covid19_Confirmed_dataset.csv" from "./Dataset" folder. 


```python
corona_dataset_csv = pd.read_csv("Datasets/covid19_Confirmed_dataset.csv")#reading dataset
corona_dataset_csv.head(20)#showing first 20 rows of the dataset
```

#### Check shape of dataframe(shows rows and columns)

```python
corona_dataset_csv.shape
```

### Deleting useless columns

```python
corona_dataset_csv.drop(["Lat","Long"],axis=1,inplace=True)
#axis=1 means we drop column
#inplace = True means it makes the updates and saves to the same dataset name
```

```python
corona_dataset_csv.head(20)
```

###  Aggregating the rows by the country

```python
aggregated_corona_dataset = corona_dataset_csv.groupby("Country/Region").sum()#referring to all data only in terms of country
```

```python
aggregated_corona_dataset.head(20)
```

```python
aggregated_corona_dataset.shape
```

### Visualizing data related to a country (Australia for example)
visualization always helps for better understanding of our data.

```python
aggregated_corona_dataset.loc["Australia"].plot()
aggregated_corona_dataset.loc["China"].plot()
aggregated_corona_dataset.loc["Spain"].plot()
plt.legend() #to give legend
```

###  Calculating a good measure 
we need to find a good measure represented as a number, describing the spread of the virus in a country. 

```python
aggregated_corona_dataset.loc["China"].plot()#plotting date vs no.of.cases for china
```

```python
aggregated_corona_dataset.loc["China"][:3].plot() #only showing data for the first three elements(dates)
```

### calculating the first derivative of the curve

```python
aggregated_corona_dataset.loc["China"].diff().plot()#plotting derivative of the curve because we aim to find the rate
```

### find maxmimum and minimum infection rate for China

```python
aggregated_corona_dataset.loc["China"].diff().max()#finding the max number of cases from derivative
```

```python
aggregated_corona_dataset.loc["China"].diff().min()#finding minimum
```

###  find maximum infection rate for all of the countries. 

```python
countries = list(aggregated_corona_dataset.index)#storing all the indices in variable countries
maxvals = []
for c in countries:
    maxvals.append(aggregated_corona_dataset.loc[c].diff().max())#appending the max values of each country index c in countries
aggregated_corona_dataset["Max infection rate"] = maxvals #adding new column called Max infection rate and it contains the values from list maxvals
```

```python
aggregated_corona_dataset.head(20)
```

### create a new dataframe with only needed column 

```python
maxrates = pd.DataFrame(aggregated_corona_dataset["Max infection rate"])#only shows max infection rate column with countries
```

```python
maxrates.head()
```
